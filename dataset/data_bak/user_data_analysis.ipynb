```python
def deal_focus_num(inputs):
    if str(inputs) == "nan":
        return 0
    spans = [
        [0, 10], [11, 50], [51, 200], [201, 500], [501, 1000], [1001, 2000], [2001, 5000], [5001, 10000], [10001, 1e12]
    ]
    for idx, (s, e) in enumerate(spans):
        if s <= inputs <= e:
            return idx + 1
    return 0

def deal_fans_num(inputs):
    if str(inputs) == "nan":
        return 0
    spans = [
        [0, 10], [11, 50], [51, 200], [201, 500], [501, 1000], [1001, 2000], [2001, 5000], [5001, 10000],
        [1e4+1, 1e5], [1e5+1, 1e6], [1e6+1, 1e7], [1e7+1, 1e8], [1e8+1, 1e9], [1e9+1, 1e12]
    ]
    for idx, (s, e) in enumerate(spans):
        if s <= inputs <= e:
            return idx + 1
    return 0

def deal_blogs_num(inputs):
    return deal_fans_num(inputs)
```

```python
excel_user_uni["focus_num"] = excel_user_uni["focus_num"].apply(lambda x: deal_focus_num(x))
excel_user_uni["fans_num"] = excel_user_uni["fans_num"].apply(lambda x: deal_fans_num(x))
excel_user_uni["blogs_num"] = excel_user_uni["blogs_num"].apply(lambda x: deal_blogs_num(x))
excel_user_uni
```
